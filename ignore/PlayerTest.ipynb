```python
from qiskit import*
circuit = QuantumCircuit(3, 3) 
#print(circuit)
circuit.h(0)
circuit.h(1)
#circuit.h(2)
#print(circuit)
#print(circuit.measure(1,1))

#circuit.measure(1,1)
from random import randint

i=0
# Player strategy in integers: R P S = 0 1 2
# R < P, 0 < 1
# P < S, 1 < 2
# S < R, 2 < 0

player = -1

#input how many times you wish for the computers to play
rounds = 1

while i < rounds:
    player = -1
    str = ""
    circuit.measure(0,0)
    simulator = Aer.get_backend('qasm_simulator')
    result = execute(circuit, backend=simulator, shots=1).result()
    counts = result.get_counts()
    if(counts.get('010')) == 1:
        i = i + 1
        str = "Rock"
        player = 0
        #print(i)
       # print(str)
    if(counts.get('000')) == 1:
        i = i + 1
       # print(i)
        str = "Paper"
        player = 1
        #print(str)
    if(counts.get('011')) == 1:
        i = i + 1
       # print(i)
        str = "Scissors"
        player = 2
       # print(str)

print("Now let's try playing against the quantum computer!")

playerinput = input("Rock, Paper, or Scissors?")
if playerinput == "Rock".casefold():
    playerinput = 0
elif playerinput == "Paper".casefold():
    playerinput = 1
elif playerinput == "Scissors".casefold():
    playerinput = 2
            
if player == 0:
    if playerinput == 1: 
        #"Paper":
        print("You win! Paper covers Rock!")
            
    elif playerinput == 2:
        print("You lose... Rock smashes Scissors!")
    
    else:
        print("You tied!")
            
elif player == 1: 
    #"Paper":
    if playerinput == 2: #"Scissors":
        print("You win! Scissors cut paper!")
            
    elif playerinput == 0:
        print("You lose... Paper covers Rock!")
        
    else:
        print("You tied!")
            
elif player == 2: 
    #"Scissors":
    if playerinput == 0: 
        #"Rock":
        print("You win! Rock smashes Scissors!")
            
    elif playerinput == 1:
        print("You lose... Scissors cut Paper!")
    
    else:
        print("You tied!")
```
